---

Функция для кодирования уровня образования: A - 7, B - 6, C - 5, D - 4, E - 3, F - 2, G - 1, также будет добавлять баллы за ступень каждого уровня: 1 - 0.8, 2 - 0.6, 3 - 0.4, 2 - 0.2. 1 - 0.

---

```python
def encode_grade_dict(x):
    res = 0.0
    
    match x[0]:
        case "A":
            res += 7
        case "B":
            res += 6
        case "C":
            res += 5
        case "D":
            res += 4
        case "E":
            res += 3
        case "F":
            res += 2
        case "G":
            res += 1
            
    match x[1]:
        case "1":
            res += 0.8
        case "2":
            res += 0.6
        case "3":
            res += 0.4
        case "4":
            res += 0.2
        case "5":
            pass
    return res

def encode_grade_df(x):
    return grades[x]
```

```python
grades = {} #словарь сопоставления уровня образования условному баллу, который будет заменять его в датафрейме
for i in sorted(df["sub_grade"].unique()):
    grades[i] = encode_grade_dict(i)
```

---

Применение написанной функции для всех значений столбца.

---

```python
df['sub_grade'] = df['sub_grade'].apply(encode_grade_df)
```

Кодирование столбца verification_status: 0 - не верифицирован, 1 - верифицирован.

```python
def enc_ver(x):
    match x:
        case "Verified":
            return 1
        case "Not Verified":
            return 0
    return 0
```

---

Функция для кодирования стажа: нет стажа - 0, < 1 года - 1, 1 год - 3, 2 года - 5, 3 год - 7, 4 года - 9, 5 лет - 11, 6 лет - 13, 7 лет - 15, 8 лет - 17, 9 лет - 19, 10+ - 25

---

```python
def encode_emp_lenght(x):
    
    match x:
        case "< 1 year":
            return 1
        case "1 year":
            return 3
        case "2 years":
            return 5
        case "3 years":
            return 7
        case "4 years":
            return 9
        case "5 years":
            return 11
        case "6 years":
            return 13
        case "7 years":
            return 15
        case "8 years":
            return 17
        case "9 years":
            return 19
        case "10+ years":
            return 25
    return 0
```

```python
df['emp_length'] = df['emp_length'].apply(encode_emp_lenght)
```

# 7. Сокращение размерности данных

---

PCA для выявления наиболее информативных признаков в данных, после чего сократим размерность данных.

---

```python
from sklearn.decomposition import PCA



pca = PCA(n_components=len(x_train.columns))



pca.fit(x_train)
```

---

Вычисление доли дисперсии, накопленной к каждой компоненте. Визуализация.

---


```python
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
```

```python
# Датафрейм для коэффициента cumulative_variance_ratio
cvr = pd.DataFrame({
    'Число компонент': range(1, len(cumulative_variance_ratio) + 1),
    'Коэффициент кумулятивной объясненной дисперсии': cumulative_variance_ratio
})

plt.figure(figsize=(10, 6))

sns.lineplot(data=cvr, 
             x='Число компонент', 
             y='Коэффициент кумулятивной объясненной дисперсии', 
             marker='o')

plt.title('Коэффициент кумулятивной объясненной дисперсии по числу компонент')
plt.xlabel('Число компонент')
plt.ylabel('Коэффициент кумулятивной объясненной дисперсии')

plt.grid(True)

plt.show()
```

---

Можно заметить, что для дисперсии 90% достаточно всего 10 компонент, а после 25 компоненты график становится почти горизонтальным. Соотвественно можно сократить размерность данных до 25 компонент.
